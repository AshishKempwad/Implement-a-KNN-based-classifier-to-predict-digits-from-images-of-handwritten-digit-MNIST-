```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
from sklearn.metrics import accuracy_score
```

```python
#Import dataset
dataset=pd.read_csv('train.csv',header=None)
test_data=pd.read_csv('test.csv',header=None) 
```

```python
#Location x is independent(label) and y is independent(vales)
x=dataset.iloc[:,:1]
y=dataset.iloc[:,1:]
z=test_data
```

```python
#Putting the first row into l and second into m
pixel=pd.DataFrame(y).to_numpy()
label=pd.DataFrame(x).to_numpy()
test_pixel=pd.DataFrame(z).to_numpy()
l=test_pixel[2]
l
```

```python
# Distance between first and all other rows
list1 =[]
for i in range(1,19999):
    m=pixel[i]
    q=label[i]
    distance = math.sqrt(sum([(a - b) ** 2 for a, b in zip(l,m)]))
    list1.append((q,distance))
    
for i in range(len(list1)):
    print(list1[i])
```

```python
#Sorting based on the second index of the list
list1.sort(key=lambda ele:ele[1])
list1
```

```python
k=5
neighbors = []
for i in range(k):
    neighbors.append(list1[i][0])
    
neighbors
```

```python
output_values = [row[-1] for row in neighbors]
prediction = max(set(output_values), key=output_values.count)
print('Predicted class is : %s' % (prediction))
```
